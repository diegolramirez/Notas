```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import ipywidgets as widg
from IPython.display import display
g = 10
```

## $l \to$ longitud (m) de la cuerda que sostiene los péndulos
## $k \to$ constante elásitca (N/m) del resorte
## $m \to$ valor de las masas (Kg) de los péndulos
## $A \to$ amplitud (m) inicial de la perturbación
## $t \to$ tiempo (s) de duración del experimento

## Modo normal 1: Ambos péndulos en fase

```python
def modo1(l,k,m,A,t):
    global g
    w0 = np.sqrt((g/l))
    w2 = np.sqrt((k/m))
    w1 = np.sqrt((w0**2 + 2*w2**2))
    T = np.linspace(0,t,500)
    x1 = A*np.cos(w1*T)
    x2 = A*np.cos(-w1*T)
    x3 = x1+x2
    fig = plt.figure(figsize=(10,8))
    plt.plot(T,x1,color="red",label="Masa 1")
    plt.plot(T,x2,label="Masa 2")
    plt.plot(T,x3,color="green",label="Combinado")
    plt.title("Movimiento de pendulo doble acoplado por un resorte")
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud (m)')
    plt.legend()
```

```python
widg.interact(modo1,
                l = widg.FloatSlider(min=0.01,max=5,step=0.01,value=0.2,),
                k = widg.FloatSlider(min=5,max=50,step=5,value=10,), 
                m = widg.FloatSlider(min=0.1,max=1,step=0.1,value=0.2,), 
                A = widg.FloatSlider(min=0.1,max=0.8,step=0.1,value=0.2,),
                t = widg.FloatSlider(min=0.5,max=8,step=0.5,value=2,),)
```

## Modo normal 2: Ambos péndulos en antifase

```python
def modo2(l,k,m,A,t):
    global g
    w0 = np.sqrt((g/l))
    w2 = np.sqrt((k/m))
    w1 = np.sqrt((w0**2 + 2*w2**2))
    T = np.linspace(0,t,500)
    x1 = A*np.cos(w1*T)
    x2 = -A*np.cos(w1*T)
    x3 = x1+x2
    fig = plt.figure(figsize=(10,8))
    plt.plot(T,x1,color="red",label="Masa 1")
    plt.plot(T,x2,label="Masa 2")
    plt.plot(T,x3,color="green",label="Combinado")
    plt.title("Movimiento de pendulo doble acoplado por un resorte")
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud (m)')
    plt.ylim(-A*1.2,A*1.2)
    plt.legend()
```

```python
widg.interact(modo2,
                l = widg.FloatSlider(min=0.01,max=5,step=0.01,value=0.2,),
                k = widg.FloatSlider(min=5,max=50,step=5,value=10,), 
                m = widg.FloatSlider(min=0.1,max=1,step=0.1,value=0.2,), 
                A = widg.FloatSlider(min=0.1,max=0.8,step=0.1,value=0.2,),
                t = widg.FloatSlider(min=0.5,max=8,step=0.5,value=2,),)
```

## Modo normal 3: Suma de los movimientos descritos por los modos normales anteriores

```python
def superposicion_modos(l,k,m,A,t):
    global g
    w0 = np.sqrt((g/l))
    w2 = np.sqrt((k/m))
    w1 = np.sqrt((w0**2 + 2*w2**2))
    T = np.linspace(0,t,500)
    x1 = A*np.cos((w1 - w0)*T/2)*np.cos((w1 + w0)*T/2)
    x2 = A*np.sin((w1 - w0)*T/2)*np.sin((w1 + w0)*T/2)
    x3 = x1+x2
    fig = plt.figure(figsize=(10,8))
    plt.plot(T,x1,color="red",label="Masa 1")
    plt.plot(T,x2,label="Masa 2")
    plt.plot(T,x3,color="green",label="Combinado")
    plt.title("Movimiento de pendulo doble acoplado por un resorte")
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud (m)')
    plt.legend()
```

```python
widg.interact(superposicion_modos,
                l = widg.FloatSlider(min=0.01,max=5,step=0.01,value=0.2,),
                k = widg.FloatSlider(min=5,max=50,step=5,value=10,), 
                m = widg.FloatSlider(min=0.1,max=1,step=0.1,value=0.2,), 
                A = widg.FloatSlider(min=0.1,max=0.8,step=0.1,value=0.2,),
                t = widg.FloatSlider(min=0.5,max=8,step=0.5,value=2,),)
```
