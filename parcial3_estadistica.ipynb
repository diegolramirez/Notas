```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def eq(x):
    return(1/(2*np.tanh(1/(2*x))))
```

```python
def ecl(x):
    return(x)
```

```python
#tamañano de imagen y letra
n = 8
```

```python
x1 = np.linspace(0,3)
y1 = eq(x1)
y2 = ecl(x1)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y1)
plt.title('Energía Cuántica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$\epsilon_q$',size=int(n/8*20))
plt.savefig("eq.png")
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y2)
plt.title('Energía Clásica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$\epsilon_{cl}$',size=int(n/8*20))
plt.savefig("ecl.png")
```

```python
def cvq(x):
    return ((1/x**2)*(np.exp(1/x)/(np.exp(1/x) - 1)**2))
```

```python
def cvcl(x):
    return (np.ones_like(x))
```

```python
y3 = cvq(x1)
y4 = cvcl(x1)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y3)
plt.title('Capacidad Calorífica Cuántica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$C_V^{q}$',size=int(n/8*20))
plt.savefig("cvq.png")
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y4)
plt.title('Capacidad Calorífica Clásica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$C_V^{cl}$',size=int(n/8*20))
plt.savefig("cvcl.png")
```

```python
def sq(x):
    return ((x + 1/2)*np.log(x + 1/2) - (x - 1/2)*np.log(x - 1/2))
```

```python
def scl(x):
    return (1 + np.log(x))
```

```python
x2 = np.linspace(1/2,3)
y5 = sq(x2)
y6 = scl(x2)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x2,y5)
plt.title('Entropía Cuántica',size=int(n/8*20))
plt.xlim(1/2,3)
plt.xlabel(r'$\frac{\epsilon}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel('s$_q$',size=int(n/8*20))
plt.savefig("sq.png")
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x2,y6)
plt.title('Entropía Clásica',size=int(n/8*20))
plt.xlim(1/2,3)
plt.xlabel(r'$\frac{\epsilon}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel('s$_{cl}$',size=int(n/8*20))
plt.savefig("sq.png")
```
