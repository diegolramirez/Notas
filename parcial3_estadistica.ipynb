```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def eq(x):
    return(1/(2*np.tanh(1/(2*x))))
```

```python
def ecl(x):
    return(x)
```

```python
#tama√±ano de imagen y letra
n = 8
```

```python
x1 = np.linspace(0,3,500)
y1 = eq(x1)
y2 = ecl(x1)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y1,c='r')
plt.title('Energ√≠a Cu√°ntica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$\epsilon_q$',size=int(n/8*20))
plt.text(1.5,1,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("eq.png",bbox_inches='tight')
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y2,c='r')
plt.title('Energ√≠a Cl√°sica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$\epsilon_{cl}$',size=int(n/8*20))
plt.text(1.5,1,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("ecl.png",bbox_inches='tight')
```

```python
def cvq(x):
    return ((1/x**2)*(np.exp(1/x)/(np.exp(1/x) - 1)**2))
```

```python
def cvcl(x):
    return (np.ones_like(x))
```

```python
y3 = cvq(x1)
y4 = cvcl(x1)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y3,c='r')
plt.title('Capacidad Calor√≠fica Cu√°ntica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$C_V^{q}$',size=int(n/8*20))
plt.text(1.5,0.2,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("cvcl.png",bbox_inches='tight')
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x1,y4,c='r')
plt.title('Capacidad Calor√≠fica Cl√°sica',size=int(n/8*20))
plt.xlim(0,3)
plt.xlabel(r'$\frac{K_BT}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel(r'$C_V^{cl}$',size=int(n/8*20))
plt.text(1.5,0.97,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("cvcl.png",bbox_inches='tight')
```

```python
def sq(x):
    return ((x + 1/2)*np.log(x + 1/2) - (x - 1/2)*np.log(x - 1/2))
```

```python
def scl(x):
    return (1 + np.log(x))
```

```python
x2 = np.linspace(1/2,3,500)
y5 = sq(x2)
y6 = scl(x2)
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x2,y5,c='r')
plt.title('Entrop√≠a Cu√°ntica',size=int(n/8*20))
plt.xlim(1/2,3)
plt.xlabel(r'$\frac{\epsilon}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel('s$_q$',size=int(n/8*20))
plt.text(1.5,0.4,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("cvcl.png",bbox_inches='tight')
```

```python
plt.figure(figsize=(8,int(n*(5/8))))
plt.plot(x2,y6,c='r')
plt.title('Entrop√≠a Cl√°sica',size=int(n/8*20))
plt.xlim(1/2,3)
plt.xlabel(r'$\frac{\epsilon}{\hbar\omega}$',size=int(n/8*20))
plt.ylabel('s$_{cl}$',size=int(n/8*20))
plt.text(1.5,0.7,r'$\hbar=\omega=K_B=N=1$',size=int(n/8*20))
plt.savefig("cvcl.png",bbox_inches='tight')
```
